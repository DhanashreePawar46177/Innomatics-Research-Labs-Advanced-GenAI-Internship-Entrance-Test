```python
# 1.User Login Check

username="admin"
password="1234"

inputUsername=input("Enter Username: ")
inputPassword=input("Enter Password: ")

if inputUsername==username and inputPassword==password:
  print("\nLogin Successful")
else:
  print("Invalid Credentials")
```

```python
# 2.Pass/Fail Analyzer

marks=[45, 78, 90, 33, 60]
passCnt=0
failCnt=0

for i in marks:
  if i>=50:
    passCnt+=1
  else:
    failCnt+=1

print("Count of Passed Students :", passCnt)
print("Count Of Failed Students :", failCnt)
```

```python
# 3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

Output_names = []

for name in names:
    Output_name = name.replace(" ", "").lower()  #Without any method
    Output_name = name.strip().lower() #With strip() method
    Output_names.append(Output_name)

print("Cleaned names:", Output_names)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for i in messages:
  print(f"Length Of '{i}' :",len(i))

  if len(i)>10:
    print(f"\nLength of '{i}' is greater than 10\n")
```

```python
# 5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
errorCnt=0

for i in logs:
  if i=="ERROR":
    errorCnt+=1

print("Count Of Error :",errorCnt)
```
